### Getting the iWeb Frequency List

```python
freq = gc.open_by_url('{###LINK-TO-FREQ-DATA###}')
            #Frequency Data from iWeb: https://www.english-corpora.org//iweb/

sheet_titles = []
for sheet in freq.worksheets():
  sheet_titles.append(sheet.title)
dict_of_sheets = {}

for sheet_title in sheet_titles:
  sheet = freq.worksheet(sheet_title)
  values = sheet.get_all_values()
  dict_of_sheets[sheet_title] = values

```

```python
df_freq = pd.DataFrame.from_records(dict_of_sheets.get('wordfreq_iWeb'))

new_header = df_freq.iloc[0] #grab the first row for the header
df_freq = df_freq[1:] #take the data less the header row
df_freq.columns = new_header #set the header row as the df header

df_freq.head(5)
```

### Get the Frequency of One Word Using df_freq

```python
#Using data frame iteration
def get_freq(word):
  word_freq=[]
  for i in range(len(df_freq)):
    if(df_freq.iloc[i, 0]==word):
      word_freq.append(word)
      word_freq.append(df_freq.iloc[i, 1])
      break
  return word_freq
```

```python
#Using dictionary key-value search
def word_freq(word):
  freq=[]
  for i in range(len(list_w)):
    if(list_w[i].get('word')==word):
      freq.append(word)
      freq.append(list_w[i].get('freq'))
      break
  return freq
```

```python
import time

start_time_ns = time.process_time_ns()
get_freq('rope')
end_time_ns = time.process_time_ns()

timer_ns = end_time_ns - start_time_ns
print(timer_ns)
```

```python
#Much quicker!
start_time_ns = time.process_time_ns()
word_freq('rope')
end_time_ns = time.process_time_ns()

timer_ns = end_time_ns - start_time_ns
print(timer_ns)
```

### Get Frequencies of Words in Sheet

```python
df2 = df.copy()
```

```python
df2.values[1,1]=get_freq(df2.iat[1,1])
```

```python
def get_freq_df(df):
  for row in range(len(df)):
    for col in range(len(df.columns)):
      df.values[row,col] = word_freq(df.iat[row,col])
  return df
```

```python
df_freq = get_freq_df(df2)
```
